### 4. Pipeline
Membuat pipepline baru dalam *collection* dengan menggunakan nama **clean_movies_nurlailiis**.

```python
imdb = list(colmovie.find({}, {'_id':0, 'imdb':1}))
imdb_list=[]
print(len(imdb))
for check in imdb:
    imdb_list.append(list(check.values())[0])
```

Perbedaan jumlah data di movies terdapat 23541 data, sedangkan di movies_initial terdapat 46014

```python
imdb_list[:5]
```

```python
pipeline = [
    {
        '$project' : {
            'title' : 1,
            'year' : 1,
            'directors' : {'$split' : ['$director', ', ']},
            'cast' : {'$split' : ['$cast', ', ']},
            'writers' : {'$split' : ['$writer', ', ']},
            'genres' : {'$split' : ['$genre', ', ']},
            'languages' : {'$split' : ['$language', ', ']},
            'countries' : {'$split' : ['$country', ', ']},
            'plot' : 1,
            # renaming the keys
            'fullplot' : '$fullplot',
            'rated' : '$rating',
            'released': {
                '$cond': {
                    'if': {'$ne': ["$released", ""]},
                    'then': {
                        '$dateFromString': {
                            'dateString': "$released"
                        }
                    },
                    'else': ""}},
            'runtime': {
                "$cond": 
                            [{'$eq': ["$runtime", ""]}, 0 ,{'$arrayElemAt':[{'$split': [ "$runtime", " min" ]}, 0]}]
                       },
            'poster' : 1,
            ### adding embeded documents ###
            ### grouping the several keys into embeded documents by one keys ###
            'imdb' : {
                'id' : '$imdbID',
                'rating' : '$imdbRating',
                'votes' : '$imdbVotes',
                },
            'metacritic':{"$cond": [{"$eq": ["$metacritic", ""]}, "$nonExistinField", "$metacritic"]},
            'awards' : {
                'win' : {'$arrayElemAt' : [{'$split': [{'$arrayElemAt':[{'$split': ['$awards','win']}, -2]}, ' ']}, -1]},
                'nomination' : {'$arrayElemAt' : [{'$split': [{'$arrayElemAt':[{'$split': ['$awards','nomination']}, -2]}, ' ']}, -1]},
                'text' : '$awards',
                },
            'type' : 1,
            'lasUpdated' : '$lastupdated',
            'num_mflix_comments': '', ## tidak ada di collection movies_initial
            'tomatoes': { ## tidak ada di collection movies_initial
                'viewer' : {
                            'rating' : '',
                            'numReviews' : '',
                            'meter' : '',
                            },
                'lasUpdated' : '',
                },
        }
    },{
        ## mencocokakan imdb di movies dan movies_initial
        '$match':{'imdb': 
                      {'$in':[dict(sorted(imdb_list[x].items(), key=lambda x: x[0])) for x in range(len(imdb_list))]}
                 }
    },
    {
        '$out' : 'clean_movies_nurlailiis'
    }
]

clear_output ()
pprint.pprint(list(client.sample_mflix.movies_initial.aggregate(pipeline)))
```

```python
db=client['sample_mflix'] #database di dbmongo
colclean=db.clean_movies_nurlailiis ## collection cluster baru
```

Terdapat beberapa variabel di *collection* movies initial atau clean_movies yang memiliki value yang berbeda dengan *collection* movies, diantaranya:

### 5. Updating Documents

```python
runtime_pat = re.compile(r'([0-9]+) min')

for movie in client.sample_mflix.clean_movies_nurlailiis.find({}):

    fields_to_set = {}
    fields_to_unset = {}

    for k,v in movie.copy().items():
        if v == "" or v == [""]:
            del movie[k]
            fields_to_unset[k] = ""

    if 'runtime' in movie:
        m = runtime_pat.match(movie['runtime']) 
        if m:
            fields_to_set['runtime'] = int(m.group(1))

    update_doc = {}
    if fields_to_set:
        update_doc['$set'] = fields_to_set
    if fields_to_unset:
        update_doc['$unset'] = fields_to_unset
    pprint.pprint(update_doc)

    db.clean_movies_nurlailiis.update_one({'_id': movie['_id']}, update_doc)
```
